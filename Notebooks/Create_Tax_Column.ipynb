```python
def set_tax_col(row):
    if row['YEAR']==2018:
        row['PRIVATE_TAX'] = 0.67
        row['SHARED_TAX'] = 0.67
        
        
        if (row['ORIGIN'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000])|(row['DESTINATION'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000]):
            row['PRIVATE_TAX']= row['PRIVATE_TAX'] + 5
            row['SHARED_TAX']=row['SHARED_TAX'] + 5

        else:
            row['PRIVATE_TAX']=row['PRIVATE_TAX']
            row['SHARED_TAX']=row['SHARED_TAX']


    elif row['YEAR']==2019:
        row['PRIVATE_TAX']= 0.72
        row['SHARED_TAX']=0.72
        
        if (row['ORIGIN'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000])|(row['DESTINATION'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000]):
            row['PRIVATE_TAX']= row['PRIVATE_TAX'] + 5
            row['SHARED_TAX']=row['SHARED_TAX'] + 5
        else:
            row['PRIVATE_TAX']=row['PRIVATE_TAX']
            row['SHARED_TAX']=row['SHARED_TAX']

    #set the shared trips tax... which is $0.65 for non downtown and $1.25 for trips to/from downtown zone
    elif row['YEAR'] == 2020:
        row['SHARED_TAX'] = 0.65
        
        #set the private trips tax... which is $1.25 for non downtown and $3 for trips to/from downtown zone
        row['PRIVATE_TAX'] = 1.25
    
        #add in the downtown zone effect
        if (row['ORIGIN'] in [17031839000,17031080202,17031839100,17031081201,17031838300,17031081401,17031081300,17031081100,17031081000,17031081900,17031081500,17031081600,17031081700,17031280100,17031842200,17031080400,17031080300,17031080201,17031080100,17031081202,17031081403,17031081402,17031320100,17031320400,17031320600,17031841900,17031281900,17031833100,17031833000,17031243400,17031243500,17031081800,17031243300])|(row['DESTINATION'] in [17031839000,17031080202,17031839100,17031081201,17031838300,17031081401,17031081300,17031081100,17031081000,17031081900,17031081500,17031081600,17031081700,17031280100,17031842200,17031080400,17031080300,17031080201,17031080100,17031081202,17031081403,17031081402,17031320100,17031320400,17031320600,17031841900,17031281900,17031833100,17031833000,17031243400,17031243500,17031081800,17031243300]):
            row['SHARED_TAX'] = row['SHARED_TAX'] + 0.60
            row['PRIVATE_TAX'] = row['PRIVATE_TAX'] + 1.75
        
        
        #add in the $5 charge for tourist hotspots
        if (row['ORIGIN'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000])|(row['DESTINATION'] in [17031980100,17031835200,17031980000,17031081402,17031330100,17031841000]):
            row['SHARED_TAX'] = row['SHARED_TAX'] + 5
            row['PRIVATE_TAX'] = row['PRIVATE_TAX'] + 5

        else:
            row['SHARED_TAX'] = row['SHARED_TAX']
            row['PRIVATE_TAX'] = row['PRIVATE_TAX']
            
    else:
        print('Bad Year')

    return row
```

```python
df = df.apply(lambda row: set_tax_col(row),axis = 1)
```
