```python
def find_distance(row, origin):
    dest_lat = row['Y']
    dest_lon = row['X']
    dest = (dest_lat, dest_lon)
    
    dist = geodesic(origin,dest).mi
    
    return dist
```

```python
df = pd.DataFrame()
for tract in tracts:
    print(tract)
    origin_lat = points[points['GEOID'] == tract]['Y'].mean()
    origin_lon = points[points['GEOID'] == tract]['X'].mean()
    origin = (origin_lat, origin_lon)
    points['DISTANCE'] = points.apply(lambda row: find_distance(row, origin), axis = 1)
    points['ORIGIN'] = tract
    df = df.append(points)
```

```python
df['DESTINATION'] = df['GEOID']
```

```python
df[['ORIGIN','DESTINATION','DISTANCE']].to_csv('C:/Workspace/TNC-Demand-Model/Outputs/Distance_Matrix.csv')
```
