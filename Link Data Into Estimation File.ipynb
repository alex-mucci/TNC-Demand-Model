```python
import pandas as pd
import geopandas as gp
```

```python
acs = pd.read_csv('C:/Workspace/TNC-Demand-Model/Outputs/ACS_2019.csv')
lehd_od = gp.read_file('C:/Workspace/TNC-Demand-Model/Outputs/OD_2019.shp')
lehd_rac = gp.read_file('C:/Workspace/TNC-Demand-Model/Outputs/RAC_2019.shp')
lehd_wac = gp.read_file('C:/Workspace/TNC-Demand-Model/Outputs/WAC_2019.shp')
```

```python
est = ridehail.merge(acs, how = 'left', left_on = 'Pickup Census Tract', right_on = 'CENSUS_TRACT', suffixes = ('','_ORIGIN'))
est = est.merge(acs, how = 'left', left_on = 'Dropoff Census Tract', right_on = 'CENSUS_TRACT', suffixes = ('','_DESTINATION'))
```

```python
est['ORIGIN_CENSUS_TRACT'] = est['Pickup Census Tract'].astype(str)
est['DESTINATION_CENSUS_TRACT'] = est['Dropoff Census Tract'].astype(str)
```

```python
est_stats = est.drop(drop_str, axis = 1)
```

```python
def find_mean(column):
    column = pd.to_numeric(column)
    output = column.mean()
    
    return output
```

```python
est_stats = est_stats.replace('-', '0')
est_stats = est_stats.replace('250,000+', '250000')
```

```python
df = pd.DataFrame()
df['Min'] = est_stats.apply(lambda column: column.min(), axis = 0)
df['Max'] = est_stats.apply(lambda column: column.max(), axis = 0)
df['Mean'] = est_stats.apply(lambda column: find_mean(column), axis = 0)
```
