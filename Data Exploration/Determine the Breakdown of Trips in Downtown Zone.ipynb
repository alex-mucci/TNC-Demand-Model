```python
import pandas as pd
```

```python
trip = pd.read_csv('D:/TNC-Demand-Model/Outputs/2019_Weekday_Monthly_Ridehail_TOD.csv')
```

```python
dz = pd.read_csv('D:/TNC-Demand-Model/Inputs/Chicago Downtown Zone/Downtown Zone Census Tracts.csv')
```

```python
dz.head()
```

```python
trip.head()
```

```python
dz['geoid10'] = dz.geoid10.astype(float)
```

```python
dest_dz = trip[trip['DESTINATION'].isin(dz.geoid10)]
```

```python
dest_dz.head()
```

```python
org_dz = trip[trip['ORIGIN'].isin(dz.geoid10)]
```

```python
org_dz.head()
```

```python
both_dz = trip[(trip['ORIGIN'].isin(dz.geoid10))&(trip['DESTINATION'].isin(dz.geoid10))]
```

```python
both_dz.head()
```

```python
neither_dz = trip[~(trip['ORIGIN'].isin(dz.geoid10))&(trip['DESTINATION'].isin(dz.geoid10))]
```

```python
neither_grouped = neither_dz[['YEAR', 'MONTH', 'TOD','SHARED_TRIPS','PRIVATE_TRIPS','ALL_TRIPS']].groupby(by = ['YEAR','MONTH','TOD']).sum()
```

```python
both_grouped = both_dz[['YEAR', 'MONTH', 'TOD','SHARED_TRIPS','PRIVATE_TRIPS','ALL_TRIPS']].groupby(by = ['YEAR','MONTH','TOD']).sum()
```

```python
org_grouped = org_dz[['YEAR', 'MONTH', 'TOD','SHARED_TRIPS','PRIVATE_TRIPS','ALL_TRIPS']].groupby(by = ['YEAR','MONTH','TOD']).sum()
```

```python
dest_grouped = dest_dz[['YEAR', 'MONTH', 'TOD','SHARED_TRIPS','PRIVATE_TRIPS','ALL_TRIPS']].groupby(by = ['YEAR','MONTH','TOD']).sum()
```

```python
neither_grouped.to_csv('D:/TNC-Demand-Model/Data Exploration/Downtown Zone Breakdown/Trips Outside Downtown Zone.csv')
```

```python
both_grouped.to_csv('D:/TNC-Demand-Model/Data Exploration/Downtown Zone Breakdown/Trips Within Downtown Zone.csv')
```

```python
org_grouped.to_csv('D:/TNC-Demand-Model/Data Exploration/Downtown Zone Breakdown/Trips Originating Within and Ending Outside Downtown Zone.csv')
```

```python
dest_grouped.to_csv('D:/TNC-Demand-Model/Data Exploration/Downtown Zone Breakdown/Trips Starting Outside and Ending Within Downtown Zone.csv')
```
